```python
import json 
emp_details=[]
n=int(input("no of employees:"))
for i in range(n):
    s={'firstname':(input(' \n firstname:-')),'lastname ':(input(' \n lastname:-')),
             'city':(input('\n city:-')),
             'CTC':int(input(' \n CTC:-')),
             'AGE':int(input('\n age:-')),
             'contact_no':int(input('\n phone_number:-'))}
    emp_details.append(s)
print(emp_details)
print(type(emp_details))
j=json.dumps(emp_details,indent=4)
with open('employee.json','w')as f: 
    json.dump(emp_details,f)
```
